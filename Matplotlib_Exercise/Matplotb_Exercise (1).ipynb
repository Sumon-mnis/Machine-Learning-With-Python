```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.plotting import scatter_matrix
```

```python
df = pd.read_csv("/content/Housing.csv")   # uploaded Kaggle dataset
df.head()
```

```python
numeric_cols = df.select_dtypes(include=np.number).columns
categorical_cols = df.select_dtypes(include="object").columns

numeric_cols, categorical_cols
```

1. Histogram

```python
plt.figure(figsize=(8,5))
plt.hist(df["price"].dropna())
plt.title("Histogram of Price")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

2. Density Plot

```python
plt.figure(figsize=(8,5))
df["price"].plot(kind="density")
plt.title("Density Plot of Price")
plt.xlabel("Price")
plt.show()
```

3. Boxplot

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df["price"])
plt.title("Boxplot of Price")
plt.show()
```

4. Violin Plot

```python
plt.figure(figsize=(8,5))
sns.violinplot(x=df["price"])
plt.title("Violin Plot of Price")
plt.show()
```

5. Scatter Plot

```python
plt.figure(figsize=(8,5))
plt.scatter(df["area"], df["price"])
plt.title("Scatter Plot: Area vs Price")
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()
```

6. Correlogram

```python
scatter_matrix(df[['price', 'area', 'bedrooms']], figsize=(9,9))
plt.show()
```

7. Heatmap

```python
plt.figure(figsize=(10,6))
sns.heatmap(df[["price","area","bedrooms","parking"]].corr(), annot=True)
plt.title("Heatmap of Selected Numeric Features")
plt.show()
```

8. Bar Chart

```python
plt.figure(figsize=(10,5))
df["furnishingstatus"].value_counts().plot(kind='bar')
plt.title("Bar Chart of Furnishing Status")
plt.xlabel("Furnishing Status")
plt.ylabel("Count")
plt.show()
```

9. Grouped Bar Chart

```python
grouped = df.groupby(["furnishingstatus","mainroad"]).size().unstack().fillna(0)

grouped.plot(kind="bar", figsize=(12,6))
plt.title("Grouped Bar Chart: Furnishing Status vs Mainroad")
plt.xlabel("Furnishing Status")
plt.ylabel("Count")
plt.show()
```

10.
Stacked Bar Chart (Explore Yourself)

```python
grouped.plot(kind="bar", stacked=True, figsize=(12,6))
plt.title("Stacked Bar Chart")
plt.xlabel("Neighborhood")
plt.ylabel("Count")
plt.show()
```
